```python
import pandas as pd
import numpy as np 

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

#COLUNA1
df_c1 = pd.read_csv('coletac1.csv') #148 -1
df_c11 = pd.read_csv('coletac1.1.csv') #156 -1
#COLUNA2
df_c2 = pd.read_csv('coletac2.csv') #167 -1
df_c22 = pd.read_csv('coletac2.1.csv') #154 -1
#COLUNA3
df_c3 = pd.read_csv('coletac3.csv') #173 -1
df_c33 = pd.read_csv('coletac3.1.csv') #193 -1
#COLUNA4
df_c4 = pd.read_csv('coletac4.csv') #164 -1
df_c44 = pd.read_csv('coletac4.1.csv') #170 -1
#COLUNA5
df_c5 = pd.read_csv('coletac5.csv') #190 -1
df_c55 = pd.read_csv('coletac5.1.csv') #174 -1
#COLUNA6    
df_c6 = pd.read_csv('coletac6.csv') #173 -1
df_c66 = pd.read_csv('coletac6.1.csv') #221 -1


arquivos = [
    'coletac1.csv', 'coletac1.1.csv',
    'coletac2.csv', 'coletac2.1.csv',
    'coletac3.csv', 'coletac3.1.csv',
    'coletac4.csv', 'coletac4.1.csv',
    'coletac5.csv', 'coletac5.1.csv',
    'coletac6.csv', 'coletac6.1.csv',
]

# Mapeamento de índices de agrupamento para os pares de arquivos
indice_agrupamento_map = {
    'coletac1.csv': 1,
    'coletac2.csv': 2,
    'coletac3.csv': 3,
    'coletac4.csv': 4,
    'coletac5.csv': 5,
    'coletac6.csv': 6,
}

dataframes_agrupados = []

# Itera sobre os pares de arquivos
for i in range(0, len(arquivos), 2):
    arquivo1 = arquivos[i]
    arquivo2 = arquivos[i + 1]

    # Carrega os dados dos arquivos CSV
    df1 = pd.read_csv(arquivo1)
    df2 = pd.read_csv(arquivo2)

    # Adiciona uma coluna com o índice de agrupamento
    indice_agrupamento = indice_agrupamento_map[arquivo1]
    df1['indice_agrupamento'] = indice_agrupamento
    df2['indice_agrupamento'] = indice_agrupamento

    # Adiciona uma coluna com o nome do arquivo
    df1['nome_arquivo'] = arquivo1
    df2['nome_arquivo'] = arquivo2

    # Concatena os dataframes
    df_agrupado = pd.concat([df1, df2], ignore_index=True)

    # Adiciona o dataframe agrupado à lista
    dataframes_agrupados.append(df_agrupado)


tabela = pd.concat(dataframes_agrupados, ignore_index=True)
tabela['Timestamp'] = pd.to_datetime(tabela['Timestamp'])
display(tabela)

tabela.to_csv('modelo.csv', index=False)


```

```python
grupos = tabela.groupby('indice_agrupamento')
primeiras_duas_linhas = []
for grupo, dados_grupo in grupos:
    primeiras_duas_linhas.append(dados_grupo.head(2))
resultado = pd.concat(primeiras_duas_linhas, ignore_index=True)
resultado.to_csv('teste.csv',index=False)
```
